```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as pt
```

```python
n = 2
A = np.random.randn(n, n)
```

```python
xs = np.random.randn(n, 1000)
```

```python
p = 1
norm_xs = np.sum(np.abs(xs)**p, axis=0)**(1/p)
norm_xs.shape
```

```python
normalized_xs = xs/norm_xs
la.norm(normalized_xs[:, 316], p)
```

```python
pt.plot(normalized_xs[0], normalized_xs[1], "o")
pt.gca().set_aspect("equal")
```

```python
A_nxs = A.dot(normalized_xs)
```

```python
pt.plot(normalized_xs[0], normalized_xs[1], "o", label="x")
pt.plot(A_nxs[0], A_nxs[1], "o", label="Ax")
pt.legend()
pt.gca().set_aspect("equal")
```

```python
norm_Axs = np.sum(np.abs(A_nxs)**p, axis=0)**(1/p)
norm_Axs.shape
```

```python
np.max(norm_Axs)
```

## Here I verify the matrix A and check the conditional number 

```python
A = np.matrix([[1.5, 0], [0, 2]])
```

```python
A_x = A@normalized_xs
```

```python
pt.plot(normalized_xs[0], normalized_xs[1], "o", label="X")
pt.plot(A_x[0], A_x[1],'o')
pt.legend()
pt.gca().set_aspect("equal")
```

```python
la.cond(A,p=1)
```

```python
A_inv = la.inv(A)
A_inv
```

```python
la.norm(A,ord=1)*la.norm(A_inv,ord=1)
```
